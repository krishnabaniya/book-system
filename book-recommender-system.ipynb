## Popularity Based Recommender System

```python
ratings_with_name = ratings.merge(books,on='ISBN')
```

```python
num_rating_df = ratings_with_name.groupby('Book-Title').count()['Book-Rating'].reset_index()
num_rating_df.rename(columns={'Book-Rating':'num_ratings'},inplace=True)
num_rating_df
```

```python
# Only select the 'Book-Title' and 'Book-Rating' columns first
avg_rating_df = ratings_with_name[['Book-Title', 'Book-Rating']].groupby('Book-Title').mean(numeric_only=True).reset_index()

# Rename the column
avg_rating_df.rename(columns={'Book-Rating': 'avg_rating'}, inplace=True)

# Display the result
avg_rating_df
```

```python
popular_df = num_rating_df.merge(avg_rating_df,on='Book-Title')
popular_df
```

```python
popular_df = popular_df[popular_df['num_ratings']>=250].sort_values('avg_rating',ascending=False).head(50)
popular_df
```

```python
print(books.columns)
```

```python
# Normalize titles to match properly
popular_df['Book-Title'] = popular_df['Book-Title'].str.strip().str.lower()
books['Book-Title'] = books['Book-Title'].str.strip().str.lower()

# Merge
merged_df = popular_df.merge(books, on='Book-Title', how='left')

# Auto-detect real column names
columns = merged_df.columns.tolist()
author_col = next((col for col in columns if 'Book-Author' in col), None)
image_col = next((col for col in columns if 'Image-URL-M' in col), None)

# Final selection
popular_df_cleaned = merged_df.drop_duplicates('Book-Title')[
    ['Book-Title', author_col, image_col, 'num_ratings', 'avg_rating']
].rename(columns={
    author_col: 'Book-Author',
    image_col: 'Image-URL-M'
})

print(popular_df_cleaned.head())
```

## Collaborative Filtering Based Recommender System

```python
x = ratings_with_name.groupby('User-ID').count()['Book-Rating'] > 200
padhe_likhe_users = x[x].index
```

```python
filtered_rating = ratings_with_name[ratings_with_name['User-ID'].isin(padhe_likhe_users)]
```

```python
y = filtered_rating.groupby('Book-Title').count()['Book-Rating']>=50
famous_books = y[y].index
```

famous_books

```python
final_ratings = filtered_rating[filtered_rating['Book-Title'].isin(famous_books)]
```

```python
pt = final_ratings.pivot_table(index='Book-Title',columns='User-ID',values='Book-Rating')
```

```python
pt.fillna(0,inplace=True)
```

```python
pt
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_scores = cosine_similarity(pt)
```

```python
similarity_scores.shape
```

```python
def recommend(book_name):
    book_name = book_name.strip().lower()

    # Normalize pt index for matching
    normalized_titles = [title.strip().lower() for title in pt.index]

    if book_name not in normalized_titles:
        return f"❌ Book '{book_name}' not found in dataset."

    index = normalized_titles.index(book_name)
    similar_items = sorted(list(enumerate(similarity_scores[index])), key=lambda x: x[1], reverse=True)[1:6]

    data = []
    for i in similar_items:
        title = pt.index[i[0]]
        temp_df = books[books['Book-Title'].str.strip().str.lower() == title.strip().lower()]
        
        if not temp_df.empty:
            item = [
                temp_df['Book-Title'].values[0],
                temp_df['Book-Author'].values[0],
                temp_df['Image-URL-M'].values[0]
            ]
            data.append(item)

    return data

```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import linear_kernel
import pandas as pd
import pickle

books = pd.read_csv("Books.csv")

# Reduce size
books = books.drop_duplicates('Book-Title').head(5000)

# Build content string
books['combined'] = (
    books['Book-Title'].fillna('') + ' ' +
    books['Book-Author'].fillna('') + ' ' +
    books['Publisher'].fillna('') + ' ' +
    books['Year-Of-Publication'].astype(str)
)


```

```python
# TF-IDF and similarity
tfidf = TfidfVectorizer(stop_words='english')
tfidf_matrix = tfidf.fit_transform(books['combined'])
content_similarity = linear_kernel(tfidf_matrix, tfidf_matrix)

# Save results
pickle.dump(content_similarity, open('content_similarity.pkl', 'wb'))
pickle.dump(books, open('books_content.pkl', 'wb'))  # optional: save reduced book list
```

```python
# Create a copy for content-based recommendations
books_content = books.drop_duplicates('Book-Title').copy()

# Assign genres to Book Titles
def assign_genre(title):
    title = title.lower()
    if "love" in title or "romance" in title:
        return "Romantic"
    elif "war" in title or "history" in title:
        return "Historical"
    elif "magic" in title or "dragon" in title or "fantasy" in title:
        return "Fantasy"
    elif "space" in title or "robot" in title or "alien" in title:
        return "Science Fiction"
    elif "murder" in title or "thriller" in title or "secret" in title:
        return "Thriller"
    else:
        return "Others"

books_content['Genre'] = books_content['Book-Title'].apply(assign_genre)

# ✅ Save this updated DataFrame
pickle.dump(books_content, open('books_content.pkl', 'wb'))
```
